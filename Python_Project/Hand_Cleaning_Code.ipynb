```python
rpm_data = [800, 1200, 9500, 1500, 2200, -500, 2800, 1100, 3100]

clean_data = []
for i in rpm_data:
    if i <= 8000 and i > 0:
        clean_data.append(i)
print(clean_data)

eco_zone = 0
power_zone = 0

for i in clean_data:
    if i <= 1500:
        eco_zone += 1
    else:
        power_zone += 1


if eco_zone > power_zone:
    print(f"ECO ZONE : {eco_zone}")
elif power_zone > eco_zone:
    print(f"POWER ZONE : {power_zone}")
else:
    print(f"ECO ZONE and POWER ZONE are equal\nECO ZONE : {eco_zone}\nPOWER ZONE : {power_zone}")
```
