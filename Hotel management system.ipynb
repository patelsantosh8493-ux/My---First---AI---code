```python
# Initializing global data structures
menu = {
    "Pizza": 250,
    "Burger": 150,
    "Pasta": 200,
    "Coffee": 80,
    "Tea": 50
}

# orders_data will be a list of dictionaries, where each dict represents an order.
# Example: [{"order_id": 1, "items": {"Pizza": 1, "Coffee": 2}}, {"order_id": 2, "items": {"Burger": 2}}]
orders_data = []
next_order_id = 1

def display_menu_func(current_menu):
    print("\n--- हमारा मेन्यू ---")
    for item, price in current_menu.items():
        print(f"{item}: ₹{price}")
    print("-------------------")

def place_order_func(current_menu, current_orders_data, order_counter):
    order_items = {}
    print(f"\nऑर्डर #{order_counter} के लिए आइटम जोड़ें:")
    while True:
        item_name = input("कौन सा आइटम चाहिए (या 'done' लिखें जब हो जाए): ").strip()
        if item_name.lower() == 'done':
            break
        if item_name in current_menu:
            try:
                quantity = int(input(f"{item_name} की कितनी मात्रा चाहिए: "))
                if quantity > 0:
                    order_items[item_name] = order_items.get(item_name, 0) + quantity
                else:
                    print("मात्रा एक धनात्मक संख्या होनी चाहिए।")
            except ValueError:
                print("अवैध मात्रा। कृपया एक संख्या दर्ज करें।")
        else:
            print(f"'{item_name}' मेन्यू में नहीं है।")

    if order_items:
        new_order = {"order_id": order_counter, "items": order_items}
        current_orders_data.append(new_order)
        print(f"\nऑर्डर #{order_counter} सफलतापूर्वक दिया गया।")
        return order_counter + 1, current_orders_data # Return updated counter and orders list
    else:
        print("कोई आइटम नहीं चुना गया।")
        return order_counter, current_orders_data # Return original counter and orders list

def get_order_by_id(order_id_to_find, current_orders_data):
    for order in current_orders_data:
        if order["order_id"] == order_id_to_find:
            return order
    return None

def display_order_details_func(order_id_to_display, current_orders_data, current_menu):
    order = get_order_by_id(order_id_to_display, current_orders_data)
    if order is None:
        print(f"ऑर्डर ID #{order_id_to_display} नहीं मिला।")
        return

    print(f"\n--- ऑर्डर ID #{order_id_to_display} विवरण ---")
    for item, quantity in order["items"].items():
        print(f"{item}: {quantity} (₹{current_menu.get(item, 'N/A')}/item)")
    print("---------------------------")

def calculate_bill_func(order_id_to_bill, current_menu, current_orders_data):
    order = get_order_by_id(order_id_to_bill, current_orders_data)
    if order is None:
        print(f"ऑर्डर ID #{order_id_to_bill} नहीं मिला।")
        return 0

    total_bill = 0
    print(f"\n--- बिल के लिए ऑर्डर ID #{order_id_to_bill} ---")
    for item, quantity in order["items"].items():
        item_price = current_menu.get(item, 0) # Use .get with default 0 if item not found (shouldn't happen if place_order checks)
        cost = item_price * quantity
        total_bill += cost
        print(f"{item} x {quantity} = ₹{cost}")
    print(f"कुल देय राशि: ₹{total_bill}")
    print("--------------------------------")
    return total_bill

# Main program loop
while True:
    print("\n--- रेस्टोरेंट मैनेजमेंट सिस्टम ---")
    print("1. मेन्यू देखें")
    print("2. ऑर्डर दें")
    print("3. बिल की गणना करें")
    print("4. सभी ऑर्डर देखें")
    print("5. बाहर निकलें")

    choice = input("अपनी पसंद दर्ज करें: ").strip()

    if choice == '1':
        display_menu_func(menu)
    elif choice == '2':
        # Update global next_order_id and orders_data
        next_order_id, orders_data = place_order_func(menu, orders_data, next_order_id)
    elif choice == '3':
        try:
            order_id_input = int(input("बिल के लिए ऑर्डर ID दर्ज करें: "))
            calculate_bill_func(order_id_input, menu, orders_data)
        except ValueError:
            print("अवैध इनपुट। कृपया एक संख्यात्मक ऑर्डर ID दर्ज करें।")
    elif choice == '4':
        if not orders_data:
            print("कोई ऑर्डर नहीं दिया गया है।")
        else:
            print("\n--- सभी सक्रिय ऑर्डर ---")
            for order_dict in orders_data:
                display_order_details_func(order_dict["order_id"], orders_data, menu)
            print("-----------------------")
    elif choice == '5':
        print("रेस्टोरेंट मैनेजमेंट सिस्टम से बाहर निकल रहा हूँ। धन्यवाद!")
        break
    else:
        print("अवैध पसंद। कृपया 1 और 5 के बीच एक संख्या दर्ज करें।")
```
